```python
from PIL import Image
import numpy as np

def image_to_rgb_arrays(image_path, target_size=(64, 64)):
    # Open the image
    img = Image.open(image_path)
    
    # Resize the image to the target size using LANCZOS filter
    img = img.resize(target_size, Image.LANCZOS)
    
    # Convert the image to RGB mode in case it's not
    img = img.convert("RGB")
    
    # Convert the image to a numpy array
    img_array = np.array(img)
    return img 

# Example usage
image_path = 'lena.jpeg'
target_size = (16, 16)  # You can try (32, 32), (64, 64), or other sizes
downscaled_img = image_to_rgb_arrays(image_path, target_size)

def grayscale(gray_values, image_size):
    # Convert the list to a numpy array and ensure it's of type uint8
    gray_array = np.array(gray_values, dtype=np.uint8)

    # Reshape the array to the specified image dimensions
    gray_array = gray_array.reshape(image_size)

    # Create a PIL Image from the 2D numpy array in grayscale mode
    img = Image.fromarray(gray_array, mode='L')  # 'L' mode is for grayscale images

    return img

def sepia(sepia_red, sepia_green, sepia_blue, image_size):
    # Convert each color list to numpy arrays and reshape to the specified image dimensions
    red_array = np.array(sepia_red, dtype=np.uint8).reshape(image_size)
    green_array = np.array(sepia_green, dtype=np.uint8).reshape(image_size)
    blue_array = np.array(sepia_blue, dtype=np.uint8).reshape(image_size)

    # Stack the red, green, and blue arrays along a third dimension to create an RGB image
    sepia_array = np.stack((red_array, green_array, blue_array), axis=-1)

    # Create a PIL Image from the 3D numpy array
    img = Image.fromarray(sepia_array, mode='RGB')  # 'RGB' mode for color images

    return img

def invert(red_values, green_values, blue_values, image_size):
    # Convert each color list to numpy arrays and reshape to the specified image dimensions
    red_array = np.array(red_values, dtype=np.uint8).reshape(image_size)
    green_array = np.array(green_values, dtype=np.uint8).reshape(image_size)
    blue_array = np.array(blue_values, dtype=np.uint8).reshape(image_size)

    # Stack the red, green, and blue arrays along a third dimension to create an RGB image
    rgb_array = np.stack((red_array, green_array, blue_array), axis=-1)

    # Create a PIL Image from the 3D numpy array
    img = Image.fromarray(rgb_array, mode='RGB')
    
    return img


gray_values = x['gray_final[7:0]']
sepia_red = y['sepia_red_final[7:0]']
sepia_green = y['sepia_green_final[7:0]']
sepia_blue = y['sepia_blue_final[7:0]']
inv_red = z['inv_red_final[7:0]']
inv_green = z['inv_green_final[7:0]']
inv_blue = z['inv_blue_final[7:0]'] 
image_size = (16, 16)  

# Display the downscaled image
downscaled_img.show()

# Create and display the grayscale image
gray_img = grayscale(gray_values, image_size)
gray_img.show()

# Create and display the sepia image
sepia_img = sepia(sepia_red, sepia_green, sepia_blue, image_size)
sepia_img.show()

# Create and display the inverted image
inv_image = invert(inv_red, inv_green, inv_blue, image_size)
inv_image.show()
```
